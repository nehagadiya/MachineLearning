**LINEAR REGRESSION**

Linear Regression is the supervised Machine Learning model in which the model finds the best fit linear line between the independent and dependent variable i.e. it finds the linear relationship between the dependent and independent variable. 

Linear Regression is of two types: Simple and Multiple. 
Simple Linear Regression is where only one independent variable is present and the model has to find the linear relationship of it with the dependent variable
In Multiple Linear Regression there are more than one independent variables for the model to find the relationship.


**Prediction Canada Capita using Univariate Linear Regression**

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df_cc = pd.read_csv("canada_capita.csv")
df_cc.head(6)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Capita")
plt.scatter(df_cc.year,df_cc.capita,color="red",marker="*")
```

```python
#creating model
lin_reg=linear_model.LinearRegression()
```

```python
#fitting the dataset in the model
lin_reg.fit(df_cc[['year']],df_cc.capita)
```

```python
#prediction for unknown input
lin_reg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.scatter(df_cc.year,df_cc.capita,color="blue",marker="*")
plt.plot(df_cc.year,lin_reg.predict(df_cc[['year']]) ,color="red")
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

**Predicting HousePrice using Multivariate Regression**

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

**Exploring dataset**

```python
df_mlr = pd.read_csv("multivariate_homeprice.csv")
df_mlr
```

```python
df_mlr.isna().sum()
```

```python
import math
bed_median = math.floor(df_mlr.bedroom.median())
bed_median
```

```python
#df_mlr will give us a series and fillna will fill the empty values of the series with the median value
df_mlr.bedroom = df_mlr.bedroom.fillna(bed_median)
```

```python
df_mlr
```

**MODEL CREATION AND FIITING THE DATASET**

```python
#model creation
multi_lin_reg = linear_model.LinearRegression()
```

```python
#fitting the dataset
multi_lin_reg.fit(df_mlr[['area','bedroom','age ']],df_mlr.price)
```

```python
multi_lin_reg.coef_
```

```python
multi_lin_reg.intercept_
```

**Prediction using the Model**

```python
multi_lin_reg.predict([[3000,3,40]])
```

```python
multi_lin_reg.predict([[2500,4,5]])
```
